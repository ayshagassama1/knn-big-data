```python
# read the file gisette_db/gisette-train.data
# and convert it to a pandas dataframe
import numpy as np
import pandas as pd
gisette_data = pd.read_csv('C:/Users/pc/Desktop/UCA/AdvanceTopML/knn-big-data/gisette-db/gisette_train.data', sep=' ', header=None)
gisette_data
```

```python
gisette_label = pd.read_csv('C:/Users/pc/Desktop/UCA/AdvanceTopML/knn-big-data/gisette-db/gisette_train.labels', sep=' ', header=None)
gisette_label
```

```python
# replace Nan with 0
gisette_data = gisette_data.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
gisette_train_data, gisette_test_data, gisette_train_label, gisette_test_label = train_test_split(gisette_data, gisette_label, test_size=0.2, random_state=42)
```

```python
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.datasets import make_blobs

# Générer des données aléatoires
#X, y = make_blobs(n_samples=1000, centers=3, random_state=42)

# Appliquer SpectralClustering
##spectral_clustering = SpectralClustering(n_clusters=20, random_state=42)
#spectral_clustering.fit(gisette_train_data)

# Appliquer KMeans pour récupérer les centres des clusters
kmeans = KMeans(n_clusters=20, random_state=42)
kmeans.fit(gisette_train_data)

# Afficher les centres des clusters
print("Centres de cluster (SpectralClustering):")
print(kmeans.cluster_centers_)
```

```python
cluster_centers_gisette_data=kmeans.cluster_centers_
cluster_centers_gisette_data
```

```python
for i in range(len(gisette_data))
   for j in range(len(cluster_centers_gisette_data))
      dist = numpy.linalg.norm(cluster_centers_gisette_data[j]-gisette_data[i])
```
